```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate random stock prices for 1 year (252 trading days)
days = 365
stock_data = np.cumsum(np.random.normal(0.001, 0.02, days))

# Calculate rolling 3-month changes


rolling_changes = np.array([100 * (stock_data[i + 63] - stock_data[i]) / stock_data[i] for i in range(len(stock_data) - 63)])

# Create a plot
plt.figure(figsize=(12, 6))
plt.plot(rolling_changes, label='3-Month Changes (%)')
plt.title('Rolling 3-Month Changes for Random Stock Data')
plt.xlabel('Trading Days')
plt.ylabel('Percentage Change')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```

```python
#plot the stock price 
plt.figure(figsize=(6,3))
plt.plot(stock_data, label='Stock Price')
plt.title('Random Stock Data')
plt.xlabel('Trading Days')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#give me the moving average of the last 90 days of the stock_data
moving_average = np.array([np.mean(stock_data[i:i+90])  for i in range(len(stock_data) - 90)])
#plot the moving average
plt.figure(figsize=(6, 3))
plt.plot(moving_average, label='90-Day Moving Average')
plt.title('90-Day Moving Average for Random Stock Data')
plt.xlabel('Trading Days')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)  
plt.show()
```
